```python
import pandas as pd
```

```python
def load_dataset():
    data = pd.read_csv('label_dataset.csv')
    return data
```

```python
tweet_df = load_dataset()
```

```python
tweet_df.head()
```

```python
df = pd.DataFrame(tweet_df)
```

```python
df
```

```python
df.drop(df.columns[[0,1,2,3]], axis=1, inplace=True)
```

```python
df
```

```python
from textblob import TextBlob
df['label'] = ''
for i,x in df.tweet.iteritems():
    label = TextBlob(x)
    df['label'][i] = label.sentiment.polarity
```

```python
print("Index: ", i ,  "label" , label.sentiment.polarity)
```

```python
df
```

```python
def polarity_to_label(x):
    if(x < 0.05):
        return 'Negative'
    if(x >= 0.05):
        return 'Positive'
df.label = df.label.apply(polarity_to_label)
```

```python
df.to_csv('textblob_dataset.csv', encoding='utf8', index=True)
```
