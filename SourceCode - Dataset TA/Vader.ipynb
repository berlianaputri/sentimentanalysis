```python
import pandas as pd
```

```python
def load_dataset():
    data = pd.read_csv('label_dataset.csv')
    return data

tweet_df = load_dataset()
df = pd.DataFrame(tweet_df)
```

```python
df
```

```python
df.drop(df.columns[[0,1,2,3]], axis=1, inplace=True)
```

```python
df
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

def VaderLabelling(tweet):
    vds = SentimentIntensityAnalyzer()
    label = vds.polarity_scores(tweet)
    
    return label['compound']
```

```python
df['label'] = df['tweet'].apply(lambda x : VaderLabelling(x))
```

```python
def sentiment_polarity(x):
    if (x >= 0.05):
        return 'Positive'
    else:
        return 'Negative'

df['sentiment'] = df['label'].apply(lambda x: sentiment_polarity(x))
```

```python
df
```

```python
df.sentiment.value_counts()
```

```python
df.to_csv('vader_dataset.csv', encoding='utf8', index=True)
```
