```python
# import library
import pandas as pd
import numpy as np
import string
import nltk
import re
```

```python
def load_dataset():
    data = pd.read_csv('merged-dataset.csv')
    return data
```

```python
tweet_df = load_dataset()
```

```python
df = pd.DataFrame(tweet_df[['text', 'timestamp']])
```

```python
df
```

```python
# remove mention
def remove_pattern(tweet, pattern):
    r = re.findall(pattern, tweet)
    for i in r :
        tweet = re.sub(i, '', tweet)
    return tweet
df['remove_user'] = np.vectorize(remove_pattern)(df['text'], "@[\w]*")
```

```python
def remove(tweet):

    # remove enter
    tweet = re.sub(r'\r\n', ' ', tweet)

    # remove angka
    tweet = re.sub('[0-9]+', '', tweet)

    # remove stock market
    tweet = re.sub(r'\$\w*', '', tweet)

    # remove RT
    tweet = re.sub(r'^RT[\s]+', '', tweet)

    # remove hashtag
    tweet = re.sub(r'#', '', tweet)

    tweet = tweet.lower()
    #Convert www.* or https?://* to URL
    tweet = re.sub('((www\.[^\s]+)|(https?://[^\s]+))','url',tweet)

    #Convert @username to AT_USER
    tweet = re.sub('@[^\s]+','AT_USER',tweet)

    #Remove additional white spaces
    tweet = re.sub('[\s]+', ' ', tweet)

    #Replace #word with word
    tweet = re.sub(r'#([^\s]+)', r'\1', tweet)
    
    #trim
    tweet = tweet.strip('\'"')
    
    return tweet

df['remove_RT'] = df['remove_user'].apply(lambda x: remove(x))
```

```python
# remove duplicate
df.sort_values("text", inplace=True)
df.drop_duplicates(subset="text", keep=False, inplace=True)
```

```python
df['remove_emoji'] = df['remove_RT'].apply(lambda x: x.encode('ascii', 'ignore').decode('ascii'))
```

```python
from nltk.corpus import stopwords
stopwords_indonesia = stopwords.words('indonesian', 'english')
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()

from nltk.tokenize import TweetTokenizer
from nltk.stem import WordNetLemmatizer
lemmatizer = nltk.stem.WordNetLemmatizer()

def clean_tweet(tweet):
    # remove angka
    tweet = re.sub('[0-9]+', '', tweet)

    # remove stock market
    tweet = re.sub(r'\$\w*', '', tweet)

    # remove RT
    tweet = re.sub(r'^RT[\s]+', '', tweet)

    # remove hashtag
    tweet = re.sub(r'#', '', tweet)

    # remove hyperlink
    tweet = re.sub(r'https?:\/\/S+', '', tweet)

    # remove coma
    tweet = re.sub(r',', '', tweet)


    # tokenize
    tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True, reduce_len=True)
    tweet_token = tokenizer.tokenize(tweet)

    tweet_clean = []
    for word in tweet_token:
        if (word not in stopwords_indonesia and word not in string.punctuation):
            # stemming word
            stem_word = stemmer.stem(word) 
            tweet_clean.append(stem_word)

    return tweet_clean

def lemmatize_text(tweet):
    tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True, reduce_len=True)
    tweet_token = tokenizer.tokenize(tweet)
    return [lemmatizer.lemmatize(w) for w in tweet_token(tweet)]
```

```python
df['tweet_clean'] = df['remove_emoji'].apply(lambda x: clean_tweet(x))
```

```python
def remove_punc(text):
    text = ' '.join([char for char in text if char not in string.punctuation])
    return text
```

```python
df['Tweet'] = df['tweet_clean'].apply(lambda x: remove_punc(x))
```
